## Import Library

```python
import cv2
import matplotlib.pyplot as plt
```

## Masukkan Gambar

```python
img = cv2.imread("uts.jpg")
```

## Konversi Gambar dari BGR to RGB 

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## Menampilkan Gambar RGB dan deteksi saluran warna

```python
red_channel = rgb[:,:,0]
green_channel = rgb[:,:,1]
blue_channel = rgb[:,:,2]

## Menampilkan masing-masing saluran warna sebagai citra grayscale
plt.figure(figsize=(15,5))

## Menampilkan citra RGB asli
plt.subplot(1, 4, 1)
plt.imshow(rgb)
plt.title('Gambar Original RGB')
plt.axis('off')

## Menampilkan channel warna merah
plt.subplot(1, 4, 2)
plt.imshow(red_channel, cmap='gray')
plt.title('Red Channel')
plt.axis('off')

## Menampilkan channel warna hijau
plt.subplot(1, 4, 3)
plt.imshow(green_channel, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

## Menampilkan channel warna biru
plt.subplot(1, 4, 4)
plt.imshow(blue_channel, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

## Menampilkan plot
plt.show()

```

## Menampilkan Histogram dari Gambar RGB dan setiap saluran nya

```python
# Hitung histogram untuk setiap saluran warna
red_hist = cv2.calcHist([red_channel], [0], None, [256], [0, 256])
green_hist = cv2.calcHist([green_channel], [0], None, [256], [0, 256])
blue_hist = cv2.calcHist([blue_channel], [0], None, [256], [0, 256])
original_hist = cv2.calcHist([rgb], [0], None, [256], [0, 256])

# Menampilkan histogram
plt.figure(figsize=(15, 10))

# Menampilkan histogram original RGB
plt.subplot(2, 4, 5)
plt.plot(original_hist, color='black')
plt.title('Histogram Gambar Original RGB')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 256])

# Menampilkan histogram saluran merah
plt.subplot(2, 4, 6)
plt.plot(red_hist, color='red')
plt.title('Histogram Red Channel')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

# Menampilkan histogram saluran hijau
plt.subplot(2, 4, 7)
plt.plot(green_hist, color='green')
plt.title('Histogram Green Channel ')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 256])

# Menampilkan histogram saluran biru
plt.subplot(2, 4, 8)
plt.plot(blue_hist, color='blue')
plt.title('Histogram Blue Channel')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

## Mengurutkan Ambang Batas Terkecil Sampai dengan Terbesar

```python
import numpy as np

# Load citra RGB asli
rgb = cv2.imread('uts.jpg')

# Konversi citra menjadi citra grayscale
gray_image = cv2.cvtColor(rgb, cv2.COLOR_BGR2GRAY)

# Tentukan nilai ambang batas
threshold_value = 0 

# Thresholding: mengatur nilai piksel di bawah ambang batas menjadi 0 (hitam) dan di atas ambang batas menjadi 255 (putih)
ret, thresholded_image = cv2.threshold(gray_image, threshold_value, 255, cv2.THRESH_BINARY)

# Menampilkan citra hasil thresholding dengan latar belakang gelap
plt.imshow(thresholded_image, cmap='gray')
plt.title('none')
plt.xlabel('')  # Menambahkan label sumbu x
plt.ylabel('')  # Menambahkan label sumbu y
plt.axis('on')  # Menampilkan sumbu x dan y
plt.show()
```
